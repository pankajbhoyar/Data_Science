## Visualization Matplotlib

```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
%matplotlib inline
```

## Plot a point

```python
plt.plot(4,3,'.')
```

```python
plt.plot(4,3,'*')
```

## Plot number of points

```python
x = np.array([2,4,6,8,10,12,14,16,18])
y = x/2

plt.figure(figsize=(10,5))
plt.scatter(x,y,c='green')
plt.show()
```

## Label the axes

```python
plt.plot([1,2,3,4])
plt.ylabel('Vertical')
plt.xlabel('Horizontal')
plt.show()
```

## Create a multiple plots with subplots

```python
names = ['PANKAJ','PRAJWAL','ANIKET','SANDIP','HIMANSHU']
score = [10,15,20,30,40]

plt.figure(figsize=(20,15))

plt.subplot(131)
plt.bar(names, score)
plt.ylabel('Vertical')
plt.xlabel('Horizontal')

plt.subplot(132)
plt.scatter(names, score)
plt.ylabel('Vertical')
plt.xlabel('Horizontal')

plt.subplot(133)
plt.plot(names, score)
plt.ylabel('Vertical')
plt.xlabel('Horizontal')

plt.suptitle('Categorical Plotting')

plt.show()
```

```python
names = ['PANKAJ','PRAJWAL','ANIKET','SANDIP','HIMANSHU']
score = [10,15,20,30,40]

plt.figure(figsize=(20,15))

plt.subplot(231)
plt.bar(names, score)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')


plt.subplot(232)
plt.scatter(names, score)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')


plt.subplot(233)
plt.plot(names, score)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')


plt.subplot(234)
plt.bar(names, score)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')

plt.suptitle('Categorical Plotting')

plt.show()
```
